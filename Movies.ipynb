<h1>Feature Engineering

```python
def classify_rating(rating):
    if rating >= 6:
        return 'High'
    elif rating >= 5:
        return 'Medium'
    else:
        return 'Low'

df['Rating_Class'] = df['Rating'].apply(classify_rating)
df["Rating_Class"].value_counts()
```

```python
df["Actor 1_Count"]=df["Actor 1"].map(df["Actor 1"].value_counts())
df["Actor 2_Count"]=df["Actor 2"].map(df["Actor 2"].value_counts())
df["Actor 3_Count"]=df["Actor 3"].map(df["Actor 3"].value_counts())
```

```python
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Actor 1"] = le.fit_transform(df["Actor 1"])
df["Actor 2"] = le.fit_transform(df["Actor 2"])
df["Actor 3"] = le.fit_transform(df["Actor 3"])
df["Director"] = le.fit_transform(df["Director"])

```

```python
df
```

```python
genre = pd.get_dummies(df["Genre"], prefix="Genre")
df = pd.concat([df, genre], axis=1)
```

```python
df.drop(columns=["Name","Genre","Rating"],axis=1,inplace=True)
```

```python
df["Rating_Class"] = le.fit_transform(df["Rating_Class"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report ,confusion_matrix
from sklearn.linear_model import LogisticRegression
x = df.drop("Rating_Class", axis=1)
y = df["Rating_Class"]
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
clf = RandomForestClassifier(random_state=42)
clf.fit(x_train_scaled, y_train)

# 5. Evaluation
y_pred = clf.predict(x_test_scaled)
print(classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
lo = LogisticRegression()
lo.fit(x_train_scaled, y_train)

# 5. Evaluation
lo_y_pred = lo.predict(x_test_scaled)
print(classification_report(y_test, lo_y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, lo_y_pred))
```
